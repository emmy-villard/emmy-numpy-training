```python
# 42. Add Elements Conditionally
# Write a NumPy program to add elements to a matrix. If an element in the matrix is 0, we will not add the element below this element.

a = np.array([[1, 1, 0, 2], [0, 3, 0, 3], [1, 0, 4, 4]])
print(a)

 # masque des zéros (à l’exception de la dernière ligne) : lignes 0 à len(a) - 1
mask = (a[:-1, :] == 0)

# on applique ce masque une ligne plus bas : lignes 1 à len(a). Et on rend égal à 0 les éléments masqués
a[1:, :][mask] = 0 

# À présent, on a transformés en 0 "l’élément en dessous des 0", donc on peut faire la somme
# (car ajouter 0 c’est comme ignorer l’élément dans une somme)
a_sum = a.sum() 

print(a_sum)
```
